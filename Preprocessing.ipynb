```python
import numpy as np 

import pandas as pd

import matplotlib.pyplot as plt

import sklearn

import os 
```

```python
import warnings
warnings.filterwarnings("ignore")
```

### Data load and simple preprocessing

```python
os.chdir('./data')
```

```python
X = pd.read_csv('train_values.csv')

X_check = pd.read_csv('test_values.csv')

y = pd.read_csv('train_labels.csv')
```

```python
mask = []
for cols in list(X.columns):
    if X[cols].dtype == 'object':
        mask.append(cols)
mask.append('building_id')
        
print('')        
print('='*200)
print('Сategorical features: \n')
print(mask)
print('='*200)
print('')

data_categorical = X.copy()[mask]
data_numerical = X.copy().drop(mask,axis = 1)
data_numerical.head()
```

```python
data_categorical
```

#### Categorical preprocessing

```python
'''Encoding categoricals'''

from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()

encoded_cat = enc.fit_transform(data_categorical.drop('building_id')).toarray()

```

```python
''' Attempting to presere the information about column names through the process of encoding'''

new_column_names = []
cols = data_categorical.columns
cats = enc.categories_

for k in range(len(cats)):
    for levels in range(len(cats[k])):
        new_column_names.append(cols[k] + '_' +  cats[k][levels])
        
```

```python
categorical_dataframe = pd.DataFrame(encoded_cat, columns = new_column_names)
categorical_dataframe['building_id'] = data_categorical['building_id']
categorical_dataframe.head()
```

#### Merging datasets

```python
preprocessed_data_train = pd.concat([data_numerical,categorical_dataframe],axis = 1)
```

```python
preprocessed_data_train.to_csv('preprocessed_train_values.csv')
```
